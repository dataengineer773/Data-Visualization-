```python
import numpy as np
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 200)
y1 = np.sin(x)
y2 = np.cos(x)

y3 = np.abs(np.sin(x))
```

```python
plt.figure(figsize=(15, 6))
plt.subplot(121)
plt.plot(x, y1, '-.' , c='green', label='sin(x)');
plt.plot(x, y2, '.', c='red', label='cos(x)');
plt.legend();
plt.title('1st plot')

plt.subplot(122)
plt.title('2nd plot')
plt.plot(x, y3, '--', c='#7D3C98');
```
